```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
def bar_chart(feature):
    survived = train[train['Survived'] == 1][feature].value_counts()
    dead = train[train['Survived'] == 0][feature].value_counts()
    print(survived)
    print('------------')
    print(dead)
    print('------------')
    df = pd.DataFrame([survived, dead])
    df.index = ['Survived', 'Dead']
    df.plot(kind = 'bar', stacked = True, figsize = (10, 5))
    
def pie_chart(feature):
    feature_ratio = train[feature].value_counts(sort = False)
    feature_size = feature_ratio.size
    feature_index = feature_ratio.index
    survived = train[train['Survived'] == 1][feature].value_counts()
    dead = train[train['Survived'] == 0][feature].value_counts()
    print(survived)
    print('------------')
    print(dead)
    print('------------')
    
    plt.plot(aspect = 'auto')
    plt.pie(feature_ratio, labels = feature_index, autopct = '%1.1f%%')
    plt.title(feature + '\s ratio in total')
    plt.show()
    
    for i, index in enumerate(feature_index):
        plt.subplot(1, feature_size + 1, i + 1, aspect = 'equal')
        plt.pie([survived[index], dead[index]], labels = ['Survived', 'Dead'], autopct = '%1.1f%%')
        plt.title(str(index) + '\s ratio')
        
    plt.show()
```

```python
train_and_test = [train, test]
print(train.shape)
print(test.shape)
```

```python
for dataset in train_and_test: # 공백으로시작하고 . 으로 끝나는 문자열 추출할때 표현
    dataset['Title'] = dataset.Name.str.extract('([A-Za-z]+)\.')
    
train.head(5)
```

```python
pd.crosstab(train['Title'], train['Sex']) # 추출한 타이틀을 가지고 있는 사람이 총 몇명인지 성별과 함께 표현한다
```

```python
for dataset in train_and_test: # 문자열을 변경해주는 replace 메소드
    dataset['Title'] = dataset['Title'].replace(['Capt', 'Col', 'Countess', 'Don','Dona', 'Dr', 'Jonkheer',
                                                 'Lady','Major', 'Rev', 'Sir'], 'Other')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
```

```python
for dataset in train_and_test: # String data 로 변경해준다
    dataset['Title'] = dataset['Title'].astype(str)
```

```python
for dataset in train_and_test: # String data 로 변경해준다
    dataset['Sex'] = dataset['Sex'].astype(str)
```

```python
# Embark 에는 Nan 이 존재함
print(train.Embarked.value_counts(dropna = False))

# Embarked Feature 의 분포를 살펴보면 S가 대부분이기 때문에 S 라고 가정해보자
for dataset in train_and_test:
    dataset['Embarked']= dataset['Embarked'].fillna('S')
    dataset['Embarked']= dataset['Embarked'].astype(str)
```

```python
# Age 도 마찬가지로 빠진값에는 일단 나머지 모든 승객의 나이의 평균을 넣어준다
# binning 이란 여러 종류의 데이터의 대해 범위를 지정해주거나 카테고리를 통해 이전보다 작은 수의 그룹으로 만드는 기법
# 단일성 분포의 왜곡을 막을 수 있지만, 이산화를 통한 데이터의 손실이라는 단점도 존재
# pd.cut()을 이용해서 같은 길이의 구간을 가지느 ㄴ다섯개의 그룹을 생성
for dataset in train_and_test:
    dataset['Age'].fillna(dataset['Age'].mean(), inplace = True)
    dataset['Age'] = dataset['Age'].astype(int)
    train['AgeBand'] = pd.cut(train['Age'], 5)
    
print (train[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean()) # Survivied ratio about Age Band
```

```python
# Age에 들어있는 값을 위에서 구한 구간에 속하도록 바꾸어줌
for dataset in train_and_test:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4
    dataset['Age'] = dataset['Age'].map( { 0: 'Child',  1: 'Young', 2: 'Middle', 3: 'Prime', 4: 'Old'} ).astype(str)
```

```python
# Fare Feature 에도 Nan 값이 존재하는데, Pclass 와 Fare 가 어느정도 연관성이 있는 것 같으므로 평균값을 그냥 넣어준다
print (train[['Pclass', 'Fare']].groupby(['Pclass'], as_index=False).mean())
print("")
print(test[test["Fare"].isnull()]["Pclass"])
```

```python
for dataset in train_and_test:
    dataset['Fare'] = dataset['Fare'].fillna(13.68)
```

```python
# SibSp & Parch Feature 형제 자매가 많을 수록 살 확률이 있기 떄문에 다음과 같이 형재들을 더해준다
for dataset in train_and_test:
    dataset['Family'] = dataset['Parch'] + dataset['SibSp']
    dataset['Family'] = dataset['Family'].astype(int)
```
